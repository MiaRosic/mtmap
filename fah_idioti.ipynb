##Učitavanje podataka

```python
Ph300 = loadmat("/content/drive/MyDrive/Colab Notebooks/Ph300.mat")
Ph500 = loadmat("/content/drive/MyDrive/Colab Notebooks/Ph500.mat")

Ph3 = Ph300['Ph300']
Ph5 = Ph500['Ph500']

ph3=Ph3.toarray()
ph5=Ph5.toarray()
data3 = pd.DataFrame(ph3)
data5 = pd.DataFrame(ph5)
```

## Grafički prikaz matrica

```python
f, x = plt.subplots(1,2, figsize=(10,10))
x[0].set_title('Ph300')
x[1].set_title('Ph500')
prvi = x[0].imshow(np.log(data3), cmap=plt.cm.get_cmap('BuGn'))
drugi = x[1].imshow(np.log(data5), cmap=plt.cm.get_cmap('PuBu'))
f.colorbar(drugi, ax=x[1], shrink=0.35)
f.colorbar(prvi, ax=x[0], shrink=0.35)
#plt.clim(-1, 1)
plt.show()
```

## Grafički prikaz svojstvenih vrijednosti

```python
f, x = plt.subplots(1,2, figsize=(8,4))
x[0].set_title('Ph300')
x[1].set_title('Ph500')
value3, vector3 = la.eig(data3)
svrijed3 = -np.sort(-abs(value3))
x[0].plot(svrijed3[0:19], 'go')
value5, vector5 = la.eig(data5)
svrijed5 = -np.sort(-abs(value5))
x[1].plot(svrijed5[0:19], 'bo')
plt.show()
```
